## Preprocessing the data

```python
#Finding the columns with missing values
null_counts=train_data.isna().sum()

for column, null in null_counts.items():
    print(f"{column} : {null}")
```

```python
#to remove the slaes price i.e. target variable from the data and store it in a different variable
target=train_data.pop('SalePrice')

#drop rows with null values from both train and test data
train_data=train_data.dropna(axis=1)
test_data=test_data.dropna(axis=1)

#finding common columns
common_columns=train_data.columns.intersection(test_data.columns)
common_columns
```

```python
train_data=train_data[common_columns]
```

```python
test_data=test_data[common_columns]
```

## Using Label Encoding for Categorical Variables

```python
categorical_columns=train_data.select_dtypes(include=['object']).columns.tolist()
# print("The categorical columns are:")
# for column in categorical_columns:
#     print(column)
```

```python
from sklearn import preprocessing
encoder=preprocessing.LabelEncoder()
for column in categorical_columns:
    train_data[column]=encoder.fit_transform(train_data[column])
    test_data[column]=encoder.fit_transform(test_data[column])
```

```python
train_data[categorical_columns]
```

## Standardizing the non-categorical columns

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
categorical_columns.append('Id')
non_categorical_columns=[column for column in train_data.columns if column not in categorical_columns]
train_data[non_categorical_columns]=scaler.fit_transform(train_data[non_categorical_columns])
test_data[non_categorical_columns]=scaler.fit_transform(test_data[non_categorical_columns])
```

## Training the model

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train=train_data
Y_train=target
```

```python
linear_reg=LinearRegression().fit(X_train,Y_train)
```

```python
Y_pred=linear_reg.predict(test_data)
```

```python
test_data['SalePrice']=Y_pred
test_data
```

```python
#Saving the results in a csv file
test_data.reset_index(inplace=True)
test_data.set_index('Id',inplace=True)
test_data[['SalePrice']].to_csv('Predicted_Prices.csv',header=['SalePrice'])
```
